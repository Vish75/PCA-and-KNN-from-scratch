```python
from sklearn.model_selection import train_test_split
X_train, y_train, X_test, y_test = train_test_split(X_train1, y_train1,test_size = 0.2, random_state= 2)
```

```python
class custom_PCA:
    def fit(self,X_train,n_comp):
        self.n_comp = n_comp
        X_meaned = X_train-np.mean(X_train,axis=0)
        cov_mat = np.cov(X_meaned, rowvar = False)
        eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
        sorted_index = np.argsort(eigen_values)[::-1]
        sorted_eigenvalue = eigen_values[sorted_index]
        sorted_eigenvectors = eigen_vectors[:,sorted_index]
        eigenvector_subset = sorted_eigenvectors[:,0:n_comp]
        X_reduced = np.dot(eigenvector_subset.transpose(),X_meaned.transpose()).transpose()
        return X_reduced
```

```python
clf = custom_PCA()
```

```python
X_transformed = clf.fit(X_train,5)
```

```python
X_transformed
```

```python
from scipy.stats import mode
```

```python
class custom_kNN:
    def dist(self,row1,row2):
        di = np.sqrt(np.sum((row1-row2)**2))
        return di
    def fit(self,X_train,k):
        labels = []
        for i in X_test:
            i_dist = []
            for j in range(len(X_train)):
                dis = self.dist(np.array(X_train[j,:]) , i)
                i_dist.append(dis)
            i_dist = np.array(i_dist)
            dist_new = np.argsort(i_dist)[:k]
            labe = y[dist_new]
            lab = mode(labe)
            lab = lab.mode[0]
            labels.append(lab)
        return labels
```

```python
new_clf = custom_kNN()
```

```python
y_pred = new_clf.fit(X_train,y_train,X_test,7)
```

```python
def acc_score(actual,predicted):
    correct = 0
    for i in range(len(actual)):
        if(actual[i]==predicted[i]):
            correct = correct +1
    return (correct/len(actual))*100.0
```

```python
print(acc_score(y_train,y_pred))
```

```python
clf2 = custom_kNN()
y_pred2 = clf2(X_transformed)
print(acc_score(y_train,y_pred2))
```
